Loading Dataset

```python
import pandas as pd

df = pd.read_csv("Twitter- datasets.csv")
df.head()
```

Checking Dataset Columns

```python
df.columns
```

```python
posts = df["description"].dropna().tolist()

len(posts)
```

Convert text to TF-IDF features

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=5000, ngram_range=(1,2))
X = vectorizer.fit_transform(posts)
```

K-Means Clustering

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=10, random_state=42)
clusters = kmeans.fit_predict(X)

print(clusters[:20])
```

CLuster Labels to dataset

```python
df["cluster"] = clusters
df[["description","cluster"]].head()
```

Calculate clustering metrics

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

sil = silhouette_score(X, clusters)
ch = calinski_harabasz_score(X.toarray(), clusters)
db = davies_bouldin_score(X.toarray(), clusters)

print("Silhouette Score:", sil)
print("Calinski-Harabasz Index:", ch)
print("Davies-Bouldin Index:", db)
```

different cluster numbers

```python
for k in range(5, 15):
    model = KMeans(n_clusters=k, random_state=42)
    labels = model.fit_predict(X)

    sil = silhouette_score(X, labels)
    print("Clusters:", k, "Silhouette Score:", sil)
```

Final K-Means Clustering Model (6 Clusters)

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=6, random_state=42)
clusters = kmeans.fit_predict(X)

df["cluster"] = clusters
df[["description","cluster"]].head()
```

Calculating metrics again

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score

sil = silhouette_score(X, clusters)
ch = calinski_harabasz_score(X.toarray(), clusters)
db = davies_bouldin_score(X.toarray(), clusters)

print("Final Silhouette Score:", sil)
print("Final Calinski-Harabasz Index:", ch)
print("Final Davies-Bouldin Index:", db)
```
