```python
# Abdul Kader Zilany (20032168)
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans

# --- New Imports for Trend Generation and Similarity ---
import yake  # <-- Added for keyword extraction
from fuzzywuzzy import fuzz  # <-- Added for fuzzy similarity
from typing import List  # <-- Added for type hints

# --- Sample dataset of social media posts ---
posts = [
    "New AI breakthrough in natural language processing!",
    "Check out the latest AI model for language tasks",
    "Football championship finals are amazing this year",
    "The football team won the finals in a thrilling match",
    "Climate change and global warming are urgent issues",
    "Global warming effects are visible in recent floods",
    "Python programming tips for beginners",
    "Learn Python with these beginner-friendly tutorials"
]

# --- Step 1: Cluster posts using KMeans ---
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(posts)

num_clusters = 3
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(X)

# Assign cluster labels to posts
df = pd.DataFrame({'post': posts, 'cluster': clusters})
print("Clustered Posts:")
print(df)

# --- Step 2: Function to generate trend name from a cluster ---
def generate_trend_name(cluster_posts: List[str], max_keywords: int = 3) -> str:
    """
    Generate a concise trend name (1-3 words) from a list of posts.
    """
    text = " ".join(cluster_posts)
    kw_extractor = yake.KeywordExtractor(top=max_keywords, stopwords=None)  # <-- Changed: Added YAKE
    keywords = kw_extractor.extract_keywords(text)

    if keywords:
        trend_name = keywords[0][0]  # Take the top keyword
        trend_name = " ".join(trend_name.split()[:3])  # Keep first 3 words
    else:
        trend_name = "Misc Trend"

    return trend_name

# --- Step 3: Function to match trend to previous trends ---
def find_similar_trend(trend_name: str, previous_trends_names: List[str], threshold: int = 70) -> str:
    """
    Compare a trend name with previous trend names and return the most similar one if above threshold.
    """
    for prev_name in previous_trends_names:
        similarity = fuzz.token_set_ratio(trend_name.lower(), prev_name.lower())  # <-- Changed: Fuzzy matching
        if similarity >= threshold:
            return prev_name
    return trend_name

# --- Step 4: Apply trend naming and similarity check ---
previous_trends = ["AI language", "Football Finals", "Climate Change"]

cluster_trends = {}
for cluster_id in df['cluster'].unique():
    cluster_posts = df[df['cluster'] == cluster_id]['post'].tolist()

    # Generate trend name (highlighted change: function usage)
    trend_name = generate_trend_name(cluster_posts)

    # Check for similarity with previous trends (highlighted change: function usage)
    trend_name = find_similar_trend(trend_name, previous_trends)

    cluster_trends[cluster_id] = trend_name

print("\nCluster Trend Names:")
for cluster_id, trend_name in cluster_trends.items():
    print(f"Cluster {cluster_id}: {trend_name}")
```
